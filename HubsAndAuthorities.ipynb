Program for å finne hubs and authority-score til nodene i en graf og avgjøre en strategisk plassering av kanter fra node 8 og 9 til enten 1,2 eller 7, slik at 7 kan komme på andreplass i å ha høyest authority score.

```python
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
#En graf hvor n8 og n9 peker på både 1,2 og 7.
G = nx.DiGraph([(3,1),(4,1),(8,1),(9,1),(5,1),(8,7),(8,2),(9,7),(9,2),(5,2),(6,2)])
nx.draw(G, with_labels="true")
nx.hits(G,max_iter=100)
```

Vi ser at n7 er på tredjeplass på høyest authoirty-score.

```python
#En graf hvor n8 bare peker på n7, og n9 peker på både n1,n2,n7.
G2 = nx.DiGraph([(3,1),(4,1),(5,1),(8,7),(9,1),(9,2),(9,7),(5,2),(6,2)])
nx.draw(G2, with_labels="true")
nx.hits(G2, max_iter=100, tol=1e-08, nstart=None, normalized=True)
```

Vi ser at n7 fremdeles er på tredjeplass. 

```python
#En graf hvor n8 og n9 peker på n1 og n7, men ikke på n2.
G3 = nx.DiGraph([(3,1),(4,1),(5,1),(8,7),(9,7),(5,2),(6,2),(9,1),(8,1)])
nx.draw(G3, with_labels="true")
nx.hits(G3, max_iter=100, tol=1e-08, nstart=None, normalized=True)
```

n7 er nå på andreplass av høyest authority-score. 

```python
#Tester en siste gang om det funker å at n8 peker på n1,n2 og n7 og n9 peker på n2 og n7. 
G4 = nx.DiGraph([(3,1),(4,1),(5,1),(8,7),(9,7),(5,2),(6,2),(9,2),(8,1),(8,2)])
nx.draw(G4, with_labels="true")
nx.hits(G4, max_iter=100, tol=1e-08, nstart=None, normalized=True)
```

n7 er på tredjeplass igjen. Kan derfor komme med konklusjonen at for at n7 skal komme opp på andreplass, er n8 og n9 peke på n7 og enten n1 eller n2.
